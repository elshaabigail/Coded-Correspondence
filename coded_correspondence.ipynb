# 1. Caesar Cipher
## 1.1 Caesar Decoder
In terms of `caesar_decoder`, the algorithm works as follows:
1. We first initialise an empty string to return the translated message.
2. We iterate through every letter in the message, and firstly we check whether the current letter exists in `punctuation` (i.e., whether the letter is a punctuation). If it is, we directly add that letter to the translated message:
```python
    else:
        translated_message += letter
```
3. If it isn't, we first find the index of the current iteration:
```python
    letter_value = alphabet.find(letter)
```
For example, the letter `d` will have a `letter_value` of 3, because it is indexed at 3 in `alphabet`.
4. We then add the index to the specified offset:
```python
    (letter_value + offset)
```
5. We modulo by 26 to ensure that the resulting addition do not go above 26 (considering we only have 26 unique characters in the entire alphabet), and instead reverts back to 1. 
```python
    (letter_value + offset) % 26
```
6. We find the value in that index within the `alphabet` variable, eventually appending it to our resulting string.

For example, the letter `w` has the index of 22 (because it is the 23rd letter of the alphabet). With an offset of 10, it becomes 22 + 10, which yields 32. Next, we modulo 32 % 26 to yield 6. Therefore, the decoded letter will be `g`, because `g` is indexed at 6 within `alphabet`.

````python
def caesar_decoder(message, offset):
  translated_message = ""
  for letter in message:
    if not letter in punctuation:
        letter_value = alphabet.find(letter)
        translated_message += alphabet[(letter_value + offset) % 26]
    else:
        translated_message += letter
  return(translated_message)
````

# 2. The Vigenère Cipher

The Vigenère cipher (French pronunciation: ​[viʒnɛːʁ]) is a method of encrypting alphabetic text by using a series of interwoven Caesar ciphers, based on the letters of a keyword. It employs a form of [polyalphabetic substitution](https://en.wikipedia.org/wiki/Polyalphabetic_cipher).

image.png

To encrypt, a table of alphabets can be used, termed a tabula recta, Vigenère square or Vigenère table. It has the alphabet written out 26 times in different rows, each alphabet shifted cyclically to the left compared to the previous alphabet, corresponding to the 26 possible Caesar ciphers. At different points in the encryption process, the cipher uses a different alphabet from one of the rows. The alphabet used at each point depends on a repeating keyword.

What this means is that opposed to having a single shift that is applied to every letter, the Vigenère Cipher has a different shift for each individual letter. ## 2.1 Keyword Encoder

1. In `punctuation`, we list all possible punctuations that we might encounter in a message. This includes (but not limited to) comma, periods, question and exclamation marks, single quote, colons and semicolons.
2.  In `message`, we represent the encoded message to be decoded
3. In `keyword` function argument, we represent the key to help us decode the message.

The algorithm works as follows:
1. We assign `space` to count how many times did we encounter elements of `punctuation` when we iterate through `message`.
2. `keyword_letter` represents the message to be coverted into keyword message
3. We iterate through `message`, where `i` represents the index of the alphabet, and `j` represents the alphabet. Hence the first elements of the iterations are:
```
('d', 0), ('f', 1), ('c', 2) ... 
```
4. We check if the alphabet being iterated is in `punctuation`. If so, skip deciphering the alphabet (i.e., add `j` to `keyword_letter`) and add 1 to `space`.
5. However, if the alphabet is not in `punctuation`: We find the index location `i`, subtracted by the number of `punctuation` occurrences (i.e., `i - space`). This is done to take into account skipped punctuations earlier on step 4.
6. We calculate the result modulo keyword length (`(i - space) % len(keyword)`)
7. We index `keyword_letter` on Step 6 results.

````python
def key_message_encoder(message, keyword):
    punctuation = ".,?'!:; " 
    space = 0
    keyword_letter = ''
    for i,j in zip(range(len(message)), message):
        if j in punctuation:
            keyword_letter += j
            space += 1
        if not j in punctuation:
            keyword_letter += keyword[((i-space)%len(keyword))]
    return(keyword_letter)        
````

````python
message = 'dfc aruw fsti gr vjtwhr wznj? vmph otis! cbx swv jipreneo uhllj kpi rahjib eg fjdkwkedhmp!'
key = key_message_encoder(message, 'friends')
print(key)
````

## 2.2 Vigenere Decoder

The algorithm works as follows:
1. We define function `vigenere_decoder` with parameter `coded_message` and `keyword_letter`
2. Inside the function, first we specify the `punctuation` and `alphabet`, where the latter is all the alphabet from a-z.
3. `translate` represents the message to be decoded
4. We iterate through `coded_message` and `keyword_letter`, where `i` represents the character in the `coded_message` and `j` represents the character in `keyword_letter`
5. We check if the character in `keyword_letter` is in punctuation. If so, we add the punctuation character to `translate`
6. However, if the alphabet is not in `punctuation`. We find the index location of `i` in alphabet using the `.find()` method of an array and assign it as `message_index`. 
7. We also find the index location of `j` in `alphabet` using the `.find()` method of an array and assign it as `keyword_index`
8. After finding the index in message and keyword, we calculate the index in alphabet using `(message_index - keyword_index) % len(alphabet)` based on the Vigenere cipher decode formula
9. We index the `alphabet` and add the result to `translate` variable

````python
def vigenere_decoder(coded_message, keyword_letter):
    punctuation = ".,?'! "
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    translate = ''
    for i,j in zip(coded_message, keyword_letter):
        if i in punctuation:
            translate += i
        else:
            message_index = alphabet.find(i)
            keyword_index = alphabet.find(j)
            translate += alphabet[(message_index - keyword_index) % len(alphabet)]
    return(translate)
````

##2.3 Vigenere Encoder

1. We define function `vigenere_encoder` with `keyword_letters` and `message` parameter. 
Note: `keyword_letters` is the `message` that was previously encoded with `keyword` (we can get it by passing key_message_encode() function into the `keyword_letters` parameter
2. Inside the function first we specify the `punctuation`, `alphabet` and `code` variable
3. `code` represent the message to be encode
4. We iterate through `keyword_letters` and `message`, where `i` represents the character in the keyword_letters and `j` represents the character in message
5. We check if the character in `keyword_letter` is in punctuation. If so, we add the `code` with the punctuation character
6. However, if the alphabet is not in punctuation. We find the index location of i in alphabet using the `.find()` method of an array and assign it as keyword_index. 
7. We also find the index location of j in alphabet using the `.find()` method of an array and assign it as message_index
8. After finding the index in message and keyword we calculate the index in alphabet using `(keyword_index - message_index) % (len(alphabet))` based on the Vigenere cypher decode formula
9. We index the alphabet and add the result to `code` variable

````python
def vigenere_encoder(keyword_letters, message):
    punctuation = ".,?'! "
    alphabet_1 = "abcdefghijklmnopqrstuvwxyz"
    code = ''
    for i,j in zip(keyword_letters, message):
        if not i in punctuation:
            keyword_index = alphabet.find(i)
            message_index = alphabet.find(j)
            code += alphabet_1[(keyword_index + message_index)%(len(alphabet_1))]
        else:
            code += i
    return(code)
````
